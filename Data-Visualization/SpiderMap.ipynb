# 蛛网图

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.font_manager import FontProperties
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 准备数据
labels = np.array([u'推进', 'KDA', u'生存', u'团战', u'发育', u'输出'])
status = [90, 70, 67, 88, 89, 99]
```

```python
# 画图数据准备，角度、状态值
angels = np.linspace(0, 2 * np.pi, len(labels) + 1, endpoint=True)
```

```python
# 首尾相连确保连线
status = np.concatenate((status, [status[0]]))
labels = np.concatenate((labels, [labels[0]]))
```

```python
# matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(angels, status, 'o-', linewidth=2)
ax.fill(angels, status, alpha=0.25)
font = FontProperties(fname=r'C:\Windows\Fonts\MSYH.ttc', size=14)
ax.set_thetagrids(angels * 180 / np.pi, labels, FontProperties=font)
plt.show()
```

```python
# 参数探索
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(angels, status, '--', linewidth=2)
ax.fill(angels, status, alpha=0.5)
font = FontProperties(fname=r'C:\Windows\Fonts\MSYH.ttc', size=12)
ax.set_thetagrids(angels * 180 / np.pi, labels, FontProperties=font)
plt.show()
```
