# 成对关系——df.describe()

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 数据准备
iris = sns.load_dataset('iris')
```

```python
# seaborn
sns.set_style('ticks')
sns.pairplot(iris,
             hue='species',
             diag_kind="kde",
             kind="scatter",
             palette="husl")
plt.show()
```
