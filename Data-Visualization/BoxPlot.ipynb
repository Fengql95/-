#  箱线图

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 数据准备
data = np.random.normal(size=(10,4))
labels = ['A', 'B', 'C','D']
```

```python
# matplotlib
plt.boxplot(data, labels=labels)
plt.show()
```

```python
# seaborn 
df = pd.DataFrame(data, columns=labels)
sns.boxplot(data = df)
plt.show()
```
