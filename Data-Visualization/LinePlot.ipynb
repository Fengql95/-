# 折线图

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 准备数据
x = [a for a in range(2010, 2020)]
y = [3,5,1,5,6,17,3,7,1,7]
```

```python
help(plt.plot)
```

```python
# matplotlib 作图
plt.plot(x,y,'o-')
plt.show()
```

```python
# seaborn 作图
df = pd.DataFrame({'x':x, 'y':y})
sns.lineplot(x='x', y='y', data=df)
plt.show()
```
