# 热力图

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# 准备数据
flights = sns.load_dataset('flights')
data = flights.pivot('year', 'month', 'passengers')
```

```python
# seaborn
sns.heatmap(data)
plt.show()
```
