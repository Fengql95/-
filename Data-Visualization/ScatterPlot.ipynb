# 散点图

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# 创建数据
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
```

```python
# matplotlib作图
plt.scatter(x, y, marker='x')
plt.show()
```

```python
plt.scatter(x, y, marker='>')
plt.show()
```

```python
plt.scatter(x, y, marker='o')
plt.show()
```

```python
# seaborn作图
df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x='x', y='y', data= df, kind='scatter')
plt.show()
```
